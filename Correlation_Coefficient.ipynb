30. The following are the heights and starting salaries of 12 law school classmates whose law school examination score were roughly the same.

| Height | Salary |
| --- | ----------- |
| 64 | 91 |
| 65 | 94 |
| 66 | 88 |
| 67 | 103 |
| 69 | 77 |
| 70 | 96 |
| 72 | 105 |
| 72 | 88 |
| 74 | 122 |
| 74 | 102 |
| 75 | 90 |
| 76 | 114 |


a) Represent these data in a scatter diagram.

b) Find the sample correlation coefficient.

**b) Correlation Coefficient**

$r = \frac{\displaystyle\sum_{i=1}^{n} (x_i - \hat{x})(y_i - \hat{y}) }{\sqrt{\displaystyle\sum_{i=1}^{n} (x_i - \hat{x})^2}{\displaystyle\sum_{i=1}^{n} (y_i - \hat{y})^2}}$

$(x_i, y_i) =$ paired values, where $i = 1,....,n$

$\hat{x}$ = sample mean of of the x values

$\hat{y}$ = sample mean of the y values

For data pair i,

$x_i - \hat{x}$ = the deviation of its x value from the sample mean

$y_i - \hat{y}$ = the deviation of its y value from the sample mean

 **Solved Mathematically**



q30.jpg






Sample mean height is $\hat{x}$

$\hat{x} = \frac{64+65+66+67+69+70+72+72+74+74+75+76}{12}$

$\hat{x} =\frac{844}{12}$

$\hat{x} = 70.33$

Sample mean salary is $\hat{y}$

$\hat{y} = \frac{91+94+88+103+77+96+105+88+122+102+90+114}{12}$

$\hat{x} =\frac{1170}{12}$

$\hat{x} = 97.5$

**Sample Correlation Coefficient**

$r = \frac{\displaystyle\sum_{i=1}^{n} (x_i - \hat{x})(y_i - \hat{y}) }{\sqrt{\displaystyle\sum_{i=1}^{n} (x_i - \hat{x})^2}{\displaystyle\sum_{i=1}^{n} (y_i - \hat{y})^2}}$

$r = \frac{272.3}{\sqrt{186.668}\sqrt{1693}}$

$r =  \frac{272.3}{13.66 \times 41.15}$

$r = 0.484$




**Solved it Programmatically** 

```
#Use the scatter() method to draw a scatter plot diagram
import matplotlib.pyplot as plt
Height = [64, 65, 66, 67, 69, 70, 72, 72, 74, 74, 75, 76]
Salary = [91, 94, 88, 103, 77, 96, 105, 88, 122, 102, 90, 114]
plt.scatter(Height, Salary)
plt.show()
```

```
# Find out mean using Pandas
#import libaries
import pandas as pd
import numpy as np

#create a dictonary datastructure
d = {'Height': [64, 65, 66, 67, 69, 70, 72, 72, 74, 74, 75, 76], 
      'Salary':[91, 94, 88, 103, 77, 96, 105, 88, 122, 102, 90, 114]}


# load data from dictionary into dataframe
df = pd.DataFrame(d)

#Display 1st five data 
print(df.head())

# Find out mean
print(df.mean())
```

```
#Another method to find out Mean and Standard deviation
import statistics

#Calculate the mean of the sample data
mean_height=statistics.mean(Height)
mean_salary= statistics.mean(Salary)
statistics.stdev(Height)
```

```
#Calculate correlation coefficient
import numpy as np
num, den_sq, den_a, den_b = 0.0, 0.0, 0.0, 0.0
for i in range(len(Height)):
  num += (Height[i]-mean_height)*(Salary[i]-mean_salary)
  den_a += np.square(Height[i]-mean_height)
  den_b += np.square(Salary[i]-mean_salary)
  den_sq = den_a * den_b
r = num/np.sqrt(den_sq)
print("The value of Correlation Coefficient is:", r)
```

33. A random group of 12 high school juniors were asked to estimate the average number of hours they study each week. The following give these hours along with the student's grade point average.

| Hours | GPA |
| --- | ----------- |
| 6 | 2.8 |
| 14 | 3.2 |
| 3 | 3.1 |
| 22 | 3.6 |
| 9 | 3.0 |
| 11 | 3.3 |
| 12 | 3.4 |
| 5 | 2.7 |
| 18 | 3.1 |
| 24 | 3.8 |
| 15 | 3.0 |
| 17 | 3.9 |

Find the sample correlation coefficient between hours reported and GPA

**Solved Mathematically**

Q32.jpg



Sample mean hours is $\hat{x}$

$\hat{x} = \frac{6+14+3+22+9+11+12+5+18+24+15+17}{12}$

$\hat{x} =\frac{156}{12}$

$\hat{x} = 13$

Sample mean GPA is $\hat{y}$

$\hat{y} = \frac{2.8+3.2+3.1+3.6+3.0+3.3+3.4+2.7+3.1+3.8+3.0+3.9}{12}$

$\hat{x} =\frac{38.9}{12}$

$\hat{x} = 3.2417$

**Sample Correlation Coefficient**

$r = \frac{\displaystyle\sum_{i=1}^{n} (x_i - \hat{x})(y_i - \hat{y}) }{\sqrt{\displaystyle\sum_{i=1}^{n} (x_i - \hat{x})^2}{\displaystyle\sum_{i=1}^{n} (y_i - \hat{y})^2}}$

$r = \frac{20.3}{\sqrt{482}\sqrt{1.5492}}$

$r =  \frac{20.3}{21.95 \times 1.24}$

$r = 0.746$


**Solved it Programmatically** 

```
#Use the scatter() method to draw a scatter plot diagram
import matplotlib.pyplot as plt
Hours = [6, 14, 3, 22, 9, 11, 12, 5, 18, 24, 15, 17]
GPA = [2.8, 3.2, 3.1, 3.6, 3.0, 3.3, 3.4, 2.7, 3.1, 3.8, 3.0, 3.9]
plt.scatter(Hours, GPA)
plt.show()
```

```
import statistics

#Calculate the mean of the sample data
mean_Hours=statistics.mean(Hours)
mean_GPA= statistics.mean(GPA)
print(mean_Hours)
print(mean_GPA)
```

```
#Calculate correlation coefficient
import numpy as np
numerator, deno_sq, deno_a, deno_b = 0.0, 0.0, 0.0, 0.0
for i in range(len(Hours)):
  numerator += (Hours[i]-mean_Hours)*(GPA[i]-mean_GPA)
  deno_a += np.square(Hours[i]-mean_Hours)
  deno_b += np.square(GPA[i]-mean_GPA)
  deno_sq = deno_a * deno_b
rcc = numerator/np.sqrt(deno_sq)
print("The value of Correlation Coefficient is:", rcc)
```
